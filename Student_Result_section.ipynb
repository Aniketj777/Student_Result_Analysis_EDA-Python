### **1) Lets check the gender distribution**

### **2) Lets check that parents educational background does affect students score or not**

```python
background = df.groupby("ParentEduc").agg({"MathScore":"mean" , "ReadingScore":"mean" , "WritingScore":"mean"})
background
```

```python
# Lets make an heatmap to get more visuals

plt.figure(figsize=(7,7))
sns.heatmap(background, annot=True)
plt.title("Relationship between Parents education and Students score")
plt.show()
```

- From the above visuals we can clearly say that students whose parents having good educational background scores good as compare to other.

- Parents having Master degree there childers having average score from 72-76
where as parenets having some high school there childers having average score from 63-54

**- So from the above chart we have conclude that the eduction of parents have a good impact on there score**

###**3)Lets Check that students score does affected on there parents marital status**

```python
Bp1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean" , "ReadingScore":"mean" , "WritingScore":"mean"})
Bp1
```

```python
plt.figure(figsize=(7,7))
sns.heatmap(Bp1, annot=True)
plt.title("Relationship between Students score and parents Marital status")
plt.show()
```
